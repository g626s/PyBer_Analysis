```python
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# When plotting x and y values, order within () matters. For reference: plt.plot(x_axis, y_axis).
plt.plot(months, savings)

plt.xlabel('Month')

plt.ylabel('Savings ($)')

plt.title('Saving Account Results')

# Show should always be the final line of code as it displays the plot. 
plt.show()
```
